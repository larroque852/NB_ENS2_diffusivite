## Ecoulement mono-dimensionnel en régime permanent en nappe captive

On souhaite exprimer la distribution spatiale de la hauteur piézométrique au sein d'une nappe captive homogène d'épaisseur constante pour un écoulement en régime permanent.
Le modèle conceptuel est donné ci-après.

<figure>
  <IMG SRC="steady1.png" WIDTH=350 ALIGN="center">
</figure>

### Utilisation de la loi de Darcy
    
Par application de la loi de Darcy, le débit $Q$ transitant à travers l'aquifère par largeur unitaire peut être exprimé par (Eq.1):
    
$$
Q = -Kb\frac{dh}{dx}
$$

La hauteur piézométrique $h$ peut alors être obtenue à partir de l'équation précédente en réécrivant (Eq.2):
    
$$
\int_{}^{} {dh} = -\int_{}^{} {\frac{Q}{Kb}dx}
$$
    et par intégration (Eq.3):
$$
h = -\frac{Q}{Kb}x + c
$$ 
    
D'après les conditions limites:
* en $ x=0$, $ h=h_0$ d'où (Eq.4):
    
$$
h_0 = -\frac{Q}{Kb} \times 0 + c
$$
    
$$
\Longrightarrow c = h_0
$$

Ce qui amène l'équation finale (Eq.5):
    
$$
h = h_0 -\frac{Q}{Kb}x
$$     

L'équation (5) relie $h$ à la position $x$ en fonction de 2 paramètres $Q$ et $Kb$ (transmissivité) et une condition limite (ici $h_0$).
Si on connaît une valeur $h_D$ en un autre point de la nappe à la distance $x=D$, l'équation (5) peut être utilisée pour estimer le débit $Q$ transitant au sein de la nappe (Eq.6):
    
$$
Q = Kb\left({ \frac{h_0 - h_D}{D}} \right)
$$
    
Et en substituant l'équation 6 dans l'équation 5, on obtient une nouvelle équation pour la hauteur piézométrique dont l'expression ne dépends plus du débit transitant au sein de la nappe (Eq.7):
   
$$
h =h_0 - \frac{h_0 - h_D}{D}x 
$$
    

*Application*

On considère un nappe captive dont les caractéristiques sont les suivantes:
 - Conductivité hydraulique K = 1e-4 m/s
 - Epaisseur b = 20 m
 
La nappe est comprise entre 2 cours d'eau, distants de *D = 500* mètres, avec lesquels elle est en relation hydraulique. Les caractéristiques hydrodynamiques des cours d'eau sont les suivantes:
 - Hauteur piézométrique du cours d'eau amont *h_0 = 20* mNG
 - Hauteur piézométrique du cours d'eau amont *h_D = 15* mNG
 
1. Tracer le profil piézométrique entre les 2 cours d'eau;
2. Quel est le débit transitant au sein de l'aquifère par largeur unitaire?


```python
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt

Kb = 1e-4*86400*20 #m2/j
h0 = 20 #mNG
hD = 15 #mNG
D = 500 #m

#profil piézométrique

x = np.arange(1,D,1)
h = h0 - ((h0 - hD)/D) * x

plt.title('Hauteur piézométrique - Nappe captive')
plt.xlabel('x (m)')
plt.ylabel('h (m NG)')

plt.grid()
plt.plot(x,h)
plt.show()

#debit transitant dans l'aquifere
Q = Kb * (h0-hD)/D

print('Le débit transitant au sein de l'' aquifère est Q = {:.3g} m3/jour'.format(Q))
```

### Utilisation de l'équation de diffusivité


Pour un écoulement monodimensionnel, l'équation de diffusivité peut s'écrire (Eq.8):

$$
\frac{d^2 h}{dx^2} = \frac{S}{Kb} \frac{dh}{dt}
$$

Pour une nappe captive, si l'aquifère présente une épaisseur constante, le produit $Kb$  qui correspond à la transmissivité est constant et est indépendant de la hauteur piézométrique $h$.

Le régime d'écoulement considéré dans le cas présent est permanent (pas de variation au cours du temps), donc (Eq.9):

$$
\frac{dh}{dt} = 0
$$

et en combinant l'équation (8) et (9), on obtient (Eq.10):

$$
\frac{d^2 h}{dx^2} = 0
$$

L'équation (10) constitue l'équation de Laplace 1D.

Pour le modèle conceptuel considéré, nous retenons les conditions limites suivantes:
- $ h = h_0 $ pour $ x = 0 $ (Eq.11)
- $ h = h_D $ pour $ x = D $ (Eq.12)

L'équation différentielle (10) peut être résolue à l'aide des conditions limites (11) et (12).
En réécrivant l'équation (10) en (Eq.13):

$$
\frac{d^2 h}{dx^2}  = 0 = \frac{d \left( {\frac{dh}{dx}} \right)}{dx}
$$

Par intégration de l'équation (13), on a (Eq.14):

$$
\int_{}^{} \frac{d \left( {\frac{dh}{dx}} \right)}{dx} = 0 \Longrightarrow \frac{dh}{dx} = C
$$

avec C constante d'intégration.

L'intégration de (14) amène (Eq.15)

$$
\int_{}^{} \frac{dh}{dx}{dx} = C \Longrightarrow h(x) = Cx + D
$$

avec D deuxième constante d'intégration.
Avec les conditions limites
- $ h = h_0 $ pour $ x = 0 $ donc $h(0) = C\times0 + D $ soit $ D = h_0 $ (Eq.16)
- $ h = h_D $ pour $ x = D $ donc $h(D) = C.D + h_0 $ soit $ D = \frac{h_D - h_0}{D}$ (Eq.17)

Finalement, par combinaison de Eq.15, Eq.16 et Eq.17, l'équation finale 1D de Laplace peut s'écrire pour le système conceptuel envisagé (Eq.18):

$$
h = \frac{h_0 - h_D}{D}x + h_0
$$

Et on retrouve la formulation obtenue à l'aide de l'équation de Darcy (Ouf !!!)

## Ecoulement mono-dimensionnel en régime permanent en nappe libre

On souhaite exprimer la distribution spatiale de la hauteur piézométrique au sein d'une nappe libre homogène pour un écoulement en régime permanent. La nappe peut être soumise à une recharge uniforme deouis la surface.
Le modèle conceptuel est donné ci-après.

<figure>
  <IMG SRC="2Drecharge.png" WIDTH=400 ALIGN="center">
</figure>

Pour un écoulement monodimensionnel en nappe libre, l'équation de Darcy peut s'écrire pour un front d'écoulement unitaire ($W = 1$) (Eq.19):

$$
q = -K\frac{dh}{dx}hW = -Kh\frac{dh}{dx}
$$

En l'absence de recharge, le débit $q$ qui transite au sein de la nappe est constant. Si on prend en compte une recharge uniforme, le débit $q$ transitant au sein de la nappe va augmenter de $dq$ au fur et à mesure de la circulation au sein de l'aquifère (cf. figure ci-après).
    
<figure>
  <IMG SRC="NL1.png" WIDTH=400 ALIGN="center">
</figure>
    
Si on considère que l'infiltration depuis la surface s'effectue au flux $N$ par unité de longueur selon la direction d'écoulement, la variation unitaire de débit $d_0$ dans l'aquifère peut s'exprimer (Eq.20):
    
$$
d_0 = Ndx
$$

et pour l'aquifère (Eq.21):
    
$$
\frac{dq}{dx} = N
$$

On peut réécrire l'équation de Darcy (Eq.19) en (Eq.22):
    
$$
q = -Kh\frac{dh}{dx} = -\frac{1}{2}K\frac{d(h^2)}{dx}
$$

Donc il vient (Eq.23):
    
$$
\frac{dq}{dx} = -\frac{1}{2}K\frac{d^2(h^2)}{dx^2}
$$
    
En reprenant l'Eq.21 et en substituant dans Eq.23, on a (Eq.24):
    
$$
-\frac{1}{2}K\frac{d^2(h^2)}{dx^2} = N
$$
    
ou encore Eq.25:
    
$$
\frac{d^2(h^2)}{dx^2} = -\frac{2N}{K}K
$$

On peut inclure le signe (-) du second terme de l'Eq.25 dans la convention du terme $N$.
La solution de l'équation différentielle Eq.25 est de la forme (Eq.26):

$$
Kh^2 + 2x^2 = C_1x + C2
$$
    
Pour le modèle conceptuel considéré, nous retenons les conditions limites suivantes:
- $ h = h_0 $ pour $ x = 0 $ (Eq.27)
- $ h = h_L $ pour $ x = L $ (Eq.28)
    
La résolution de l'équation Eq.62 pour les conditions limites Eq.27 et Eq.28 amène à l'équation de la surface piézométrique suivante:
 
$$
h = \sqrt{h_0^2 + \frac{h_L^2-h_0^2}{L}x + \frac{N}{K}(L-x)x}
$$

*Application*

On considère une nappe libre de conductivité hydraulique K = 1e-4 m/s.
La nappe est comprise entre 2 cours d'eau, distants de *D = 500* mètres, avec lesquels elle est en relation hydraulique. Cette nappe libre est soumise à une recharge efficace $N = 0.4$ mm d'eau.

Les caractéristiques hydrodynamiques des cours d'eau sont les suivantes:
 - Hauteur piézométrique du cours d'eau amont *h_0 = 20* mNG
 - Hauteur piézométrique du cours d'eau aval *h_L = 15* mNG
 

1. Tracer le profil piézométrique entre les 2 cours d'eau en considérant une recharge efficace $N = 0.4$ mm d'eau;
2. Dans le cas où la recharge est nulle, quel est le débit transitant au sein de l'aquifère à mi-chemin et juste avant le cours d'eau constituant l'exutoire, pour un front d'écoulement unitaire?

*Correction question 2:*

- Débit transitant au sein de l'aquifère à mi-chemin ($x=l_m$)

Equation de Darcy:
$$
q = -K\frac{dh}{dx}hW = -Kh\frac{dh}{dx}
$$

Après ré-arrangement pour intégration:
$$
qdx = -Khdh
$$

$$
\int_{0}^{l_m} {q}\mathrm{d}x = \int_{h_0}^{h_{l_m}} {-Kh}\mathrm{d}h
$$

$$
ql_m  = -K \left({\frac{h^2_{l_m} - h^2_0}{2}} \right)
$$

$$
q  = - \frac{K}{2l_m} \left({h^2_{l_m} - h^2_0} \right)
$$

- Débit transitant au sein de l'aquifère à l'exutoire ($x=D$): la démonstration est identique pour des bornes différentes. Au final:

$$
q  = - \frac{K}{2L} \left({h^2_{D} - h^2_0} \right)
$$



```python
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt

K = 1e-4*86400 #m/j
h0 = 20 #mNG
hL = 15 #mNG
N = 0.00 #m H2O
L = 500 #m

#calcul de hmid=h(L/2)
x = L/2
hmid = np.sqrt(h0 **2 + ((hL ** 2 - h0 ** 2)/(L)) * x + N/K* (L-x)*x)
print(hmid)
qmid = -(K/(2*x))*(hmid ** 2 - h0 ** 2)

qex = -(K/(2*L))*(hL ** 2 - h0 ** 2)

print('Le débit transitant au sein de l'' aquifère à mi-chemin est Q = {:.3g} m3/jour'.format(qmid))
print('Le débit transitant au sein de l'' aquifère à l''exutoire est Q = {:.3g} m3/jour'.format(qex))

```
